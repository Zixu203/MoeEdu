```python
import pandas as pd
df = pd.read_excel('/content/drive/MyDrive/Shared/edu_bigdata_imp.xlsx')
```

```python
df_selected_drop_no_dupl = df[['city_code','classification', 'action_name', 'resource_name', 'video_item_sn', 'month', 'grade', 'review_start_time', 'prac_score_rate', 'review_sn']]
df_selected_drop_no_dupl.loc[df_selected_drop_no_dupl['classification'].isnull(),'class_NaN'] = True
df_selected_drop_no_dupl.loc[df_selected_drop_no_dupl['classification'].notnull(),'class_NaN'] = False
df_selected_drop_no_dupl.loc[df_selected_drop_no_dupl['prac_score_rate'].isnull(),'prac_NaN'] = True
df_selected_drop_no_dupl.loc[df_selected_drop_no_dupl['prac_score_rate'].notnull(),'prac_NaN'] = False
```

```python
df_selected_drop_no_dupl
```

```python
df_input = pd.DataFrame()
```

```python
df_data06 = df[["PseudoID", "city_code", "review_sn", "review_start_time"]].drop_duplicates()

df_data06 = df_data06.sort_values(by=['PseudoID', 'city_code', "review_start_time"])

df_data06 = df_data06.dropna(how='any',axis=0)

city_dict = {}

for i in df_data06['city_code'].unique().tolist():
    city_dict[i] = (df_data06.loc[df_data06.city_code == i])
```

```python
city_review_total = [[0, 0] for _ in range(0,22)]

for i in city_dict:
    city_review_total[i][0] = i
    city_review_total[i][1] = len(city_dict[i])

# for i in city_review_total:
#     try:
#         city_review_total.remove([0,0])
#     except:
#         None

df_input['city_count'] = df_selected_drop_no_dupl['city_code'].apply(lambda x: city_review_total[x][1])

hot_time_df = pd.DataFrame()
city_hotest_time = [[0, 0] for _ in range(0,22)]
for i in city_dict:
    hot_time_list = [0] * 24
    for j in city_dict[i]["review_start_time"]:
        hot_time_list[j.hour] += 1
    hot_time_df[i] = hot_time_list
hot_time_df = hot_time_df.reindex(sorted(hot_time_df.columns), axis=1)

for i in hot_time_df.columns:
    city_hotest_time[i] = hot_time_df[i].idxmax()

print(city_hotest_time)

df_input['hot_time'] = df_selected_drop_no_dupl['city_code'].apply(lambda x: city_hotest_time[x])
```

```python
df_selected_drop_no_dupl['classification'].isnull()
```

```python
import numpy as np

df_input['math_onehot'] = [0] * 101730
df_input['nature_onehot'] = [0] * 101730
df_input['society_onehot'] = [0] * 101730
df_input['technology_onehot']= [0] * 101730
df_input['english_onehot'] = [0] * 101730
df_input['lang_onehot'] = [0] * 101730
df_input['art_onehot'] = [0] * 101730
df_input['else_onehot'] = [0] * 101730

for i in range(len(df_selected_drop_no_dupl['classification'])):
  if df_selected_drop_no_dupl["class_NaN"][i]:
    df_input['else_onehot'][i] = 1
  else :
    str_x = df_selected_drop_no_dupl['classification'][i]
    if '數學' in str_x : df_input['math_onehot'][i] = 1
    elif '自然' in str_x : df_input['nature_onehot'][i]  = 1
    elif '社會' in str_x : df_input['society_onehot'][i] = 1
    elif '科技' in str_x : df_input['technology_onehot'][i] = 1
    elif '英語' in str_x : df_input['english_onehot'][i] = 1
    elif '語文' in str_x : df_input['lang_onehot'][i]  = 1
    elif '藝術' in str_x : df_input['art_onehot'][i] = 1
    else: df_input['else_onehot'][i] = 1
```

```python
df_input['video_item_sn'] = df_selected_drop_no_dupl['video_item_sn']
```

```python
df_input['city_code'] = df_selected_drop_no_dupl['city_code']
# df_input['month'] = df_selected_drop_no_dupl['month']
# df_input['grade'] = df_selected_drop_no_dupl['grade']
df_input['prac_score_rate'] = df_selected_drop_no_dupl['prac_score_rate']
```

```python
df_input['score_low_and_medium'] = df_selected_drop_no_dupl['prac_score_rate'].apply(lambda x: 1 if x <= 70 else 0)
df_input['score_high'] = df_selected_drop_no_dupl['prac_score_rate'].apply(lambda x: 1 if x > 70 else 0)
df_input['count_high'] = df_input['city_count'].apply(lambda x: 1 if x <= 100 else 0)
df_input['count_low'] = df_input['city_count'].apply(lambda x: 1 if x > 100 else 0)
```

```python
select_col = ['math_onehot', 'nature_onehot', 'society_onehot', 'technology_onehot', 'english_onehot', 'lang_onehot','art_onehot', 'else_onehot', 'score_low_and_medium', 'score_high', 'count_low', 'count_high', 'hot_time']
train = df_input[select_col].to_numpy()
test = df_input['city_code'].to_numpy()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split
var_train, var_test, res_train, res_test = train_test_split(train, test,test_size = 0.3)

decision_tree = tree.DecisionTreeClassifier()
decision_tree = decision_tree.fit(var_train, res_train)
```

```python
score = decision_tree.score(var_test, res_test)
score
```
